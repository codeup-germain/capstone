```python
def get_match(df, match):
    
    d = {}
    
    for value, player in enumerate(match):
        
        champ_id = match[value]['championId']
        champ_name = match[value]['championName']
        kills = match[value]['kills']
        deaths = match[value]['deaths']
        assists = match[value]['assists']
        lane = match[value]['lane']
        team_position = match[value]['teamPosition']
        individual_position = match[value]['individualPosition']
        participant_id = match[value]['participantId']
        
        d['champ_id_'+str(value+1)] = champ_id
        d['champ_name_'+str(value+1)]= champ_name
        d['kills_'+str(value+1)]= kills
        d['deaths_'+str(value+1)]= deaths
        d['assists_'+str(value+1)]= assists
        d['lane_'+str(value+1)]= lane
        d['team_position_'+str(value+1)]= team_position 
        d['individual_position_'+str(value+1)]= individual_position
        d['participant_id_'+str(value+1)]= participant_id
    
    df = df.append(d, ignore_index=True)
    return df
```

```python
df = pd.DataFrame()
for match in match_ids:
    query = f'lol/match/v5/matches/{match}?api_key={api_key}'
    response = requests.get(url+query)
    game = response.json()
    match = game['info']['participants']
    df = get_match(df, match)
```
